```python
df1['duration_call_to']=df1.duration*df1.call_to
df1['duration_call_in']=df1.duration*df1.call_in
df1['duration_call_to_miss']=df1.duration*df1.call_to_miss
df1['duration_call_in_miss']=df1.duration*df1.call_in_miss
df1.member_phone=df1.member_phone.astype('int64').astype('string').astype('object')
df1
```

```python
# chuẩn bị dữ liệu tính pagranks
to1=(df1.call_in.astype('bool')*df1.member_phone.astype('int64').astype('string').astype('object')).astype('object')
to2=(df1.call_to.astype('bool')*df1.phone.astype('object')).astype('object')
to3=(df1.call_in_miss.astype('bool')*df1.member_phone.astype('int64').astype('string').astype('object')).astype('object')
to4=(df1.call_to_miss.astype('bool')*df1.phone.astype('object')).astype('object')

df1['to']=to1+to2+to3+to4
```

```python
in1=(df1.call_to.astype('bool')*df1.member_phone.astype('int64').astype('string').astype('object')).astype('object')
in2=(df1.call_in.astype('bool')*df1.phone.astype('object')).astype('object')
in3=(df1.call_to_miss.astype('bool')*df1.member_phone.astype('int64').astype('string').astype('object')).astype('object')
in4=(df1.call_in_miss.astype('bool')*df1.phone.astype('object')).astype('object')

df1['in']=in1+in2+in3+in4
df1.info()
```

```python
rank=df1[['to','in']]
rank['weight']=df1.duration
rank
```

```python
import networkx as nx
G = nx.from_pandas_edgelist(rank,source='to',target='in',edge_attr=['weight'],create_using=nx.DiGraph)

pr = nx.pagerank(G)

pageranks = pd.DataFrame(pr.items(), columns=["title", "pagerank"])
pageranks.pagerank=pageranks.pagerank*100000
```

```python
pageranks.to_csv('pageranks.csv')
pageranks.info()
```

```python
def addF(a):
    a['avg_in_contact']=a.in_contact/a.sum_call
    a['avg_duration']=0.0
    for i in a.index:
        if (a.call_to[i]+a.call_in[i])>0:
            a['avg_duration'][i]=a.duration[i]/(a.call_to[i]+a.call_in[i])
    a['avg_duration_call_to']=a.duration_call_to/(a.call_to+0.0001+a.call_to_miss)
    a['avg_duration_call_in']=a.duration_call_in/(a.call_in+0.0001+a.call_in_miss)
    a['avg_success']=a.success/a.sum_call
    return a
a=addF(a)
m=addF(m)
m
```
