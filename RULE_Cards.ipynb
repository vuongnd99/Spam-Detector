```python
tmp_emails=[]
tmp_phones=[]
tmp_website=[]
tmp_name=[]
tmp_name_email=[]
tmp_name_web=[]
for i in df.index:
    w=df.words[i]
    email=re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", w)
    tmp_name_email.append(re.findall(r'(.*?)@',' '.join([elem for elem in email])))
    tmp_emails.append(email)
    phones=(re.findall(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]', w))
    tmp_phones.append(phones)
    website=(re.findall(r'(?:http://?)+[a-zA-Z0-9-_./]+[a-zA-Z]|(?:www.?)+[a-zA-Z-_0-9.]+[a-zA-Z]+',w))
    tmp_website.append(website)
    tmp_name_web.append(re.findall(r'www.(.*?)\.',' '.join([elem for elem in website])))
#     tmp_name.append(re.findall(re.compile(r'([A-Z]+(?: [A-Z]\.)+z[A-Z]+)'),w))/

# df['name']=tmp_name


# df['address']=0
df['website']=tmp_website
df['name_web']=tmp_name_web
df['emails']=tmp_emails
df['name_email']=tmp_name_email
df['name_web']=tmp_name_web
df['phones']=tmp_phones
df.sample(60)
```
