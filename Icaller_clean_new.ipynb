```python
a=[0]*20
print(a[0])
for i in notlabel.phone:
    for j in range(0,20):
        if len(str(i)) == j:
#             print(i)
            a[j] = a[j]+1
print(a)
```

```python
phone = []
for i in notlabel.phone:
    if (len(str(i)) == 1)  or (len(str(i)) == 2) or (len(str(i)) == 3)  or (len(str(i)) == 5) :
        phone.append(i)
print(len(phone))
```

```python
yes = [0]*len(phone)
tam = {'phone' : phone, 'yes' : yes}
phone_in = pd.DataFrame(tam)
phone_in = phone_in.set_index('phone')
```

```python
phone_dict = phone_in.transpose().to_dict()
```

```python
tam =[]
for i in range(len(notlabel)):
    try: 
        tam.append(phone_dict[(notlabel.phone[i])]['yes'])
    except:
        tam.append(1)
        
```

```python
notlabel['in_phone'] = tam
notlabel = notlabel.drop((notlabel.loc[notlabel.in_phone == 0].index),axis = 0)
```
